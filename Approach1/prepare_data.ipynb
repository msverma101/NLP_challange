```python
import pandas as pd
pd.set_option('display.width', 1000)


# Load the dataset
df = pd.read_csv("training.csv", index_col=0, header=0, names=["idx", "company", "output", "input"])

# Shuffle the DataFrame
df = df.sample(frac=1).reset_index(drop=True)

# Display the first few rows
print(df.head())

print("items insgesamt: " + str(len(df)))
print()
df.output.value_counts().plot(kind='bar');
```

```python
import pandas as pd

for subset in ["training.csv", "validation.csv"]:
    # Read CSV with specified columns
    df = pd.read_csv(subset, index_col=0, header=0, names=["idx", "company", "output", "input"])
    
    # Set 'instruction' column to a fixed string
    df["instruction"] = "Detect the sentiment of the tweet."
    df["input"] = df["company"] + " " + df["input"]
 
    # Drop the 'company' column
    df = df.drop(columns=["company"])
    
    # Save to JSON with proper formatting
    df.to_json(subset.replace(".csv", ".json"), orient="records", lines=False)
```
